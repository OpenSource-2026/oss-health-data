````python
import os
import time
import requests
import pandas as pd
from datetime import datetime, timezone
from dotenv import load_dotenv

load_dotenv()
token = os.getenv("GITHUB_TOKEN")
HEADERS={
    "Accept": "application/vnd.github+json"
}
if token:
    HEADERS["Authorization"]=f"Bearer {token}"
BASE_URL = "https://api.github.com"
FULL_NAME = "pandas-dev/pandas"
repo_url = f"{BASE_URL}/repos/{FULL_NAME}"
````

````python
def fetch_json(url, params=None, headers=HEADERS, sleep_sec=0.2):
    res = requests.get(url, headers=headers, params=params)
    res.raise_for_status()
    time.sleep(sleep_sec)
    return res.json()
````

JSON 문자열 → Python dict/list로 파싱됨 → 구조 기반 순회

````python
from pathlib import Path

def is_github_api_url(value):
    return isinstance(value, str) and value.startswith("https://api.github.com/")

def is_template_url(value):
    return isinstance(value, str) and ("{" in value and "}" in value)

def save_progress(rows, error_log, save_prefix="walk_json_progress"):
    pd.DataFrame(rows).to_csv(f"{save_prefix}_rows.csv", index=False)
    pd.DataFrame(error_log).to_csv(f"{save_prefix}_errors.csv", index=False)

def walk_json(
    obj,
    parent_path="",
    depth=0,
    max_depth=1,
    visited_urls=None,
    error_log=None,
    rows=None,
    verbose=True,
    save_every=100,
    counter=None,
    save_prefix="walk_json_progress"
):
    if visited_urls is None:
        visited_urls = set()
    if error_log is None:
        error_log = []
    if rows is None:
        rows = []
    if counter is None:
        counter = {"n": 0}

    # depth 제한
    if depth > max_depth:
        return rows, error_log

    def append_row(row):
        rows.append(row)
        counter["n"] += 1

        if save_every and counter["n"] % save_every == 0:
            save_progress(rows, error_log, save_prefix)
            if verbose:
                print(f"[checkpoint] saved {counter['n']} rows")

    if isinstance(obj, dict):
        for key, value in obj.items():
            current_path = f"{parent_path}.{key}" if parent_path else key

            append_row({
                "path": current_path,
                "key": key,
                "dtype": type(value).__name__,
                "example_value": str(value)[:120],
                "is_github_api_url": is_github_api_url(value),
                "fetch_status": None,
                "error_message": None,
                "depth": depth,
            })

            if isinstance(value, (dict, list)):
                walk_json(
                    value,
                    parent_path=current_path,
                    depth=depth + 1,
                    max_depth=max_depth,
                    visited_urls=visited_urls,
                    error_log=error_log,
                    rows=rows,
                    verbose=verbose,
                    save_every=save_every,
                    counter=counter,
                    save_prefix=save_prefix
                )

            elif (
                depth < max_depth
                and is_github_api_url(value)
                and not is_template_url(value)
                and value not in visited_urls
            ):
                visited_urls.add(value)

                try:
                    sub_data = fetch_json(value)

                    append_row({
                        "path": current_path + ".__api_fetch__",
                        "key": "__api_fetch__",
                        "dtype": type(sub_data).__name__,
                        "example_value": str(value)[:120],
                        "is_github_api_url": True,
                        "fetch_status": "success",
                        "error_message": None,
                        "depth": depth,
                    })

                    if sub_data is not None:
                        walk_json(
                            sub_data,
                            parent_path=current_path + ".__api__",
                            depth=depth + 1,
                            max_depth=max_depth,
                            visited_urls=visited_urls,
                            error_log=error_log,
                            rows=rows,
                            verbose=verbose,
                            save_every=save_every,
                            counter=counter,
                            save_prefix=save_prefix
                        )

                except requests.exceptions.HTTPError as e:
                    msg = f"HTTPError: {e}"
                    error_log.append({
                        "path": current_path,
                        "url": value,
                        "error_type": "HTTPError",
                        "error_message": str(e),
                        "depth": depth,
                    })

                    append_row({
                        "path": current_path + ".__api_error__",
                        "key": "__api_error__",
                        "dtype": "error",
                        "example_value": str(value)[:120],
                        "is_github_api_url": True,
                        "fetch_status": "failed",
                        "error_message": msg,
                        "depth": depth,
                    })

                    if verbose:
                        print(f"[HTTPError] {current_path} -> {value}")
                        print(f"  {e}")

                except Exception as e:
                    msg = f"{type(e).__name__}: {e}"
                    error_log.append({
                        "path": current_path,
                        "url": value,
                        "error_type": type(e).__name__,
                        "error_message": str(e),
                        "depth": depth,
                    })

                    append_row({
                        "path": current_path + ".__api_error__",
                        "key": "__api_error__",
                        "dtype": "error",
                        "example_value": str(value)[:120],
                        "is_github_api_url": True,
                        "fetch_status": "failed",
                        "error_message": msg,
                        "depth": depth,
                    })

                    if verbose:
                        print(f"[Error] {current_path} -> {value}")
                        print(f"  {msg}")

    elif isinstance(obj, list):
        for idx, item in enumerate(obj):
            current_path = f"{parent_path}[{idx}]"

            append_row({
                "path": current_path,
                "key": f"[{idx}]",
                "dtype": type(item).__name__,
                "example_value": str(item)[:120],
                "is_github_api_url": is_github_api_url(item),
                "fetch_status": None,
                "error_message": None,
                "depth": depth,
            })

            if isinstance(item, (dict, list)):
                walk_json(
                    item,
                    parent_path=current_path,
                    depth=depth + 1,
                    max_depth=max_depth,
                    visited_urls=visited_urls,
                    error_log=error_log,
                    rows=rows,
                    verbose=verbose,
                    save_every=save_every,
                    counter=counter,
                    save_prefix=save_prefix
                )

            elif (
                depth < max_depth
                and is_github_api_url(item)
                and not is_template_url(item)
                and item not in visited_urls
            ):
                visited_urls.add(item)

                try:
                    sub_data = fetch_json(item)

                    append_row({
                        "path": current_path + ".__api_fetch__",
                        "key": "__api_fetch__",
                        "dtype": type(sub_data).__name__,
                        "example_value": str(item)[:120],
                        "is_github_api_url": True,
                        "fetch_status": "success",
                        "error_message": None,
                        "depth": depth,
                    })

                    if sub_data is not None:
                        walk_json(
                            sub_data,
                            parent_path=current_path + ".__api__",
                            depth=depth + 1,
                            max_depth=max_depth,
                            visited_urls=visited_urls,
                            error_log=error_log,
                            rows=rows,
                            verbose=verbose,
                            save_every=save_every,
                            counter=counter,
                            save_prefix=save_prefix
                        )

                except requests.exceptions.HTTPError as e:
                    msg = f"HTTPError: {e}"
                    error_log.append({
                        "path": current_path,
                        "url": item,
                        "error_type": "HTTPError",
                        "error_message": str(e),
                        "depth": depth,
                    })

                    append_row({
                        "path": current_path + ".__api_error__",
                        "key": "__api_error__",
                        "dtype": "error",
                        "example_value": str(item)[:120],
                        "is_github_api_url": True,
                        "fetch_status": "failed",
                        "error_message": msg,
                        "depth": depth,
                    })

                    if verbose:
                        print(f"[HTTPError] {current_path} -> {item}")
                        print(f"  {e}")

                except Exception as e:
                    msg = f"{type(e).__name__}: {e}"
                    error_log.append({
                        "path": current_path,
                        "url": item,
                        "error_type": type(e).__name__,
                        "error_message": str(e),
                        "depth": depth,
                    })

                    append_row({
                        "path": current_path + ".__api_error__",
                        "key": "__api_error__",
                        "dtype": "error",
                        "example_value": str(item)[:120],
                        "is_github_api_url": True,
                        "fetch_status": "failed",
                        "error_message": msg,
                        "depth": depth,
                    })

                    if verbose:
                        print(f"[Error] {current_path} -> {item}")
                        print(f"  {msg}")

    # 마지막에도 저장
    if depth == 0:
        save_progress(rows, error_log, save_prefix)
        if verbose:
            print(f"[done] total rows={len(rows)}, total errors={len(error_log)}")

    return rows, error_log
````

### Nested GitHub API Flatten 결과 정리

---

### 정리 목적

현재 `walk_json`을 통해 flatten한 결과에는 repository 자체의 health를 직접 설명하는 정보와,
nested endpoint에서 파생된 개별 객체 정보가 혼합되어 있다.

최종 목표는 **repository 단위의 health feature table 구성**이므로,
모든 flattened row를 그대로 유지하는 것은 비효율적이다.

따라서 각 데이터는 아래 3가지 기준으로 구분한다.

* 그대로 유지
* 집계 후 사용
* 제거

---

### 최종 분류 기준표

| 데이터 그룹             | 예시 path / 형태                                                        | 처리 방식   | 이유               | 활용 방식              |
| ------------------ | ------------------------------------------------------------------- | ------- | ---------------- | ------------------ |
| Repository root 통계 | stargazers_count, forks_count, subscribers_count, open_issues_count | 유지      | repo-level 정량 정보 | 직접 feature 사용      |
| Repository 상태 flag | has_issues, has_wiki, archived, disabled, allow_forking             | 유지      | 운영 상태 설명         | governance feature |
| 시간 정보              | created_at, updated_at, pushed_at                                   | 유지      | 활동성/신선도 반영       | time-based feature |
| endpoint URL       | contributors_url, languages_url 등                                   | 집계 후 사용 | 추가 데이터 source    | fetch 후 feature 생성 |
| contributor 객체     | contributors_url.**api**[i]                                         | 집계 후 사용 | 개별 user row는 불필요 | 분포/집중도 계산          |
| contributor 프로필    | login, id, avatar_url 등                                             | 제거      | 식별자 수준 정보        | 사용 안 함             |
| language raw       | languages_url.**api**.Python 등                                      | 집계 후 사용 | 언어 구성 의미 있음      | entropy / 비율       |
| tag raw            | tags_url.**api**[i]                                                 | 집계 후 사용 | release 정보 포함    | maturity feature   |
| tag 내부 URL         | zipball_url 등                                                       | 제거      | 다운로드 링크          | 의미 없음              |
| event raw          | events_url.**api**[i]                                               | 집계 후 사용 | 활동 유형 중요         | type count         |
| event actor        | actor.login 등                                                       | 제거      | 개인 식별 정보         | 의미 없음              |
| deployment raw     | deployments_url.**api**[i]                                          | 집계 후 사용 | 존재 여부만 중요        | count/flag         |
| stargazer raw      | stargazers_url.**api**[i]                                           | 제거      | count로 대체 가능     | 제거                 |
| subscriber raw     | subscribers_url.**api**[i]                                          | 제거      | count로 충분        | 제거                 |
| URL/템플릿            | commits_url, contents_url 등                                         | 제거      | navigation용      | 제거                 |
| 식별자                | id, node_id                                                         | 제거      | 의미 없음            | 제거                 |

### **[Tags]**
> git/github의 버전 표식

**=> features to make**

* **num_tags**

  * 전체 tag 개수
  * 프로젝트의 릴리즈 이력 규모를 나타냄

* **stable_tag_ratio** (정식 릴리즈 중심 프로젝트인지, 실험 및 개발 단계 비중이 큰지)

  * stable: `v3.0.2`, `v2.3.1`
  * unstable/prerelease: `dev`, `rc`, `alpha`, `beta`
  * 해석:

    * 높음 → 안정적인 릴리즈 중심
    * 낮음 → 개발/실험 단계 비중 큼

* **prerelease_tag_ratio** (개발 중간 단계 태그가 얼마나 많은지)

  * 예: `v3.1.0.dev0`, `v3.0.0rc1`
  * 해석:

    * 높음 → active development 상태
    * 낮음 → 안정화된 프로젝트

* **latest_tag_is_stable**

  * 가장 최신 tag가 stable인지 여부
  * 해석:

    * True → 최근 안정 릴리즈 존재
    * False → 현재 개발 진행 중일 가능성

* **latest_tag_is_prerelease**

  * 최신 tag가 dev/rc/beta인지 여부
  * 해석:

    * True → 아직 release 전 단계
    * False → 안정 버전 상태

* **semver_tag_ratio** (버전 규칙 사용 여부)

  * 예: `vMAJOR.MINOR.PATCH` → `v3.0.2`
  * 해석:

    * 높음 → 버전 관리 체계적
    * 낮음 → 비정형/임의 태그 사용

* **num_major_versions / num_minor_versions** (optional)

  * major: `v1.x.x`, `v2.x.x`, `v3.x.x`
  * 해석:

    * 프로젝트의 진화 단계 추정 가능

### **[Events]**

**=> features to make**

* **num_events**

  * 전체 event 개수
  * 최근 활동량 자체를 나타냄
  * 해석:

    * 높음 → 활발한 프로젝트
    * 낮음 → 활동 저조 / 정체 가능성
  * 사용 데이터:

    * 각 이벤트 객체 1개 = 1 event
    * → 전체 리스트 길이 (`len(events)`)

* **num_unique_event_types**

  * 등장한 event type 종류 수
  * 해석:

    * 높음 → 다양한 상호작용 (issue, PR, comment 등)
    * 낮음 → 특정 활동에 치우침
  * 사용 데이터:

    * `event["type"]` 값
    * → set으로 unique 개수 계산

* **IssuesEvent_ratio** (이슈 관리 비중)

  * 전체 event 중 IssuesEvent 비율
  * 해석:

    * 높음 → 이슈 기반 관리 활발
  * 사용 데이터:

    * `event["type"] == "IssuesEvent"` 개수 / 전체 이벤트 수

* **IssueCommentEvent_ratio** (토론/피드백 비중)

  * 전체 event 중 IssueCommentEvent 비율
  * 해석:

    * 높음 → 커뮤니케이션 활발
  * 사용 데이터:

    * `event["type"] == "IssueCommentEvent"` 개수 / 전체 이벤트 수
    * (예: 실제 데이터에서 `"type": "IssueCommentEvent"` 확인됨 )

* **PullRequestEvent_ratio** (코드 기여 흐름)

  * 전체 event 중 PullRequestEvent 비율
  * 해석:

    * 높음 → PR 기반 협업 구조
  * 사용 데이터:

    * `event["type"] == "PullRequestEvent"` 개수 / 전체 이벤트 수

* **PushEvent_ratio** (직접 개발 반영 비중)

  * 전체 event 중 PushEvent 비율
  * 해석:

    * 높음 → core contributor 중심 개발
  * 사용 데이터:

    * `event["type"] == "PushEvent"` 개수 / 전체 이벤트 수

* **WatchEvent_ratio** (외부 관심 비중)

  * 전체 event 중 WatchEvent 비율
  * 해석:

    * 높음 → 사용자 관심 증가
  * 사용 데이터:

    * `event["type"] == "WatchEvent"` 개수 / 전체 이벤트 수

* **ForkEvent_ratio** (재사용/확장 관심 비중)

  * 전체 event 중 ForkEvent 비율
  * 해석:

    * 높음 → 프로젝트 활용 및 확장 활발
  * 사용 데이터:

    * `event["type"] == "ForkEvent"` 개수 / 전체 이벤트 수

* **dominant_event_type**

  * 가장 많이 발생한 event type
  * 해석:

    * 프로젝트의 주요 활동 성격 요약
  * 사용 데이터:

    * `event["type"]` 빈도 카운트 → 최빈값

* **dominant_event_ratio**

  * 가장 많은 event type이 전체에서 차지하는 비율
  * 해석:

    * 높음 → 특정 활동에 집중
    * 낮음 → 다양한 활동 존재
  * 사용 데이터:

    * max(event_type_count) / 전체 이벤트 수

* **event_type_entropy** (활동 다양성)

  * event type 분포의 엔트로피
  * 해석:

    * 높음 → 다양한 활동 균형
    * 낮음 → 특정 활동 편중
  * 사용 데이터:

    * 각 event type 비율 p
    * → `-Σ(p * log p)`

* **has_IssuesEvent / has_PullRequestEvent / has_IssueCommentEvent (optional)**

  * 특정 event 존재 여부 (0/1)
  * 해석:

    * 기능 존재 여부 확인
  * 사용 데이터:

    * `event["type"]` 포함 여부 (set membership)

* **recent_event_density (optional)**

  * 이벤트 발생 밀도 (시간 기반)
  * 해석:

    * 높음 → 매우 활발
    * 낮음 → 느린 개발
  * 사용 데이터:

    * `event["created_at"]`
    * → max - min 시간 차 / 이벤트 수
    * (예: `"created_at": "2026-04-03T10:47:53Z"` 존재 )


### **[Deployments]**
> 이 repo의 특정 코드(ref/sha)를 어떤 환경(environment)에 배포하려고 만든 배포 기록

> deployments 데이터는 repository의 배포 이력과 코드 버전(ref) 기반 운영 여부를 나타내며, 배포 존재 여부, 배포 대상(ref) 다양성, 그리고 최근 배포 시점을 통해 프로젝트의 릴리즈/운영 활성도를 보조적으로 측정할 수 있다.

> deployments는 “얼마나 자주, 어떤 코드 기준으로 배포되고 있는가”를 보는 최소한의 운영 지표다.


**=> features to make**

* **num_deployments**

  * 전체 deployment 개수
  * 프로젝트가 배포 이력을 가지고 있는지, 얼마나 자주 배포되는지의 최소한의 지표
  * 해석:

    * 높음 → 배포 활동 존재
    * 낮음 → 배포 이력 거의 없음
  * 사용 데이터:

    * deployment 객체 개수
    * → 전체 리스트 길이 (`len(deployments)`)

* **has_deployments**

  * deployment 기록 존재 여부 (0/1)
  * 해석:

    * True → 배포 프로세스 존재
    * False → 배포 기록 없음
  * 사용 데이터:

    * deployment 리스트가 비어있는지 여부

* **num_unique_refs**

  * deployment에 사용된 서로 다른 ref 개수
  * (브랜치 또는 태그 기준)
  * 해석:

    * 높음 → 다양한 코드 버전이 배포됨
    * 낮음 → 제한된 ref만 배포됨
  * 사용 데이터:

    * `deployment["ref"]`
    * unique 개수 계산

* **tag_based_deployment_ratio**

  * 전체 deployment 중 tag 기반 배포 비율
  * (예: `ref = v3.0.2`)
  * 해석:

    * 높음 → 릴리즈(버전) 기반 배포 구조
    * 낮음 → 브랜치 기반 배포 중심
  * 사용 데이터:

    * `deployment["ref"]`
    * → `v\d+\.\d+\.\d+` 같은 버전 패턴 여부 판별 후 비율 계산

* **deployment_recency_days**

  * 가장 최근 deployment 기준으로 현재까지 지난 시간 (일 단위)
  * 해석:

    * 낮음 → 최근에도 배포 중 (활성 상태)
    * 높음 → 배포 오래 없음
  * 사용 데이터:

    * `deployment["created_at"]`
    * → 가장 최신 값 기준으로 현재 시각과 차이 계산
